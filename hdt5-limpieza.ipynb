```python
import pandas as pd
import numpy as np
from nltk.corpus import stopwords
import nltk
#nltk.download()
```

```python
df = pd.read_csv('data/train.csv')
df.head()
```

```python
df['text']  = df['text'].str.lower()
df['text'].replace(to_replace=stopwords.words('english'), value='', inplace=True)
df['text']
```

```python
df.replace(to_replace=r'http\S+', value='', regex=True, inplace=True)
```

```python
#df['hashtags'] = df['text'].str.findall('#(\w+)') 
df['text'].replace(to_replace = r'#(\w+)', value='', inplace=True, regex=True) 
df['text'].replace(to_replace = r'@(\w+)', value='', inplace=True, regex=True) 
```

```python
df['text'].replace(to_replace = r'\d+', value='', inplace=True, regex=True) 
```

```python
df['text'].replace(to_replace=r"""[!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~]""", value='', inplace=True, regex=True)
```

```python
#[i for i in df['text'].str.findall(r'\d+') if i != [] and '911' in i]
```

```python
tokenized = []
for i, e in enumerate(df['text']):
    tokenized.append(nltk.tokenize.word_tokenize(e, language="english"))
```
