```python
import pandas as pd
import json
from tensorflow import keras
from tensorflow.keras import layers
import tensorflow as tf
import nltk
import re
import numpy as np
from nltk.util import ngrams
```

```python
with open('indexes.sav', 'r') as f:
    indexes = json.loads(f.read())
indexes
```

```python
CNN = keras.models.load_model('save_at_25.h5')
```

```python
def clasificador(s):
    s = s.lower()
    s = re.sub(r'#(\w+)', '', s)
    s = re.sub(r'http\S+', '', s)
    s = re.sub(r'\d+', '', s)
    s = re.sub(r"""[!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~]""", '', s)
    n_gram_onehot = np.zeros(shape=(1, len(indexes)), dtype=np.float32)
    print(n_gram_onehot.shape)
    tokens = nltk.tokenize.word_tokenize(s, language="english")
    for n in range(1, 4):
        for j in ngrams(tokens, n):
            j = str(j)
            if j in indexes.keys():
                n_gram_onehot[0][indexes[j]] = 1.0

    pred = CNN.predict(n_gram_onehot)

    return ('No accidente','Accidente')[np.argmax(pred)]
```

```python
clasificador('grand theft auto')
```
