```python
import json
from tensorflow import keras
from tensorflow.keras import layers
import tensorflow as tf
import nltk
import re
import numpy as np
from nltk.util import ngrams
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
with open('hdt5indexes.sav', 'r') as f:
    indexes = json.loads(f.read())
indexes
```

```python
CNN = keras.models.load_model('hdt5_save_at_40.h5')
```

```python
analyzer = SentimentIntensityAnalyzer()
def clasificador(s):
    s = s.lower()
    s = re.sub(r'#(\w+)', '', s)
    s = re.sub(r'http\S+', '', s)
    s = re.sub(r'\d+', '', s)
    s = re.sub(r"""[!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~]""", '', s)
    n_gram_onehot = np.zeros(shape=(1, len(indexes)+1), dtype=np.float32)
    print(n_gram_onehot.shape)
    tokens = nltk.tokenize.word_tokenize(s, language="english")
    for n in range(1, 3):
        for j in ngrams(tokens, n):
            j = str(j)
            if j in indexes.keys():
                n_gram_onehot[0][indexes[j]] = 1.0
    n_gram_onehot[0][len(indexes)] = analyzer.polarity_scores(s)['compound'] 
    pred = CNN.predict(n_gram_onehot)

    return ('No accidente','Accidente')[np.argmax(pred)]
```

```python
clasificador('This earthquake Allah')
```
